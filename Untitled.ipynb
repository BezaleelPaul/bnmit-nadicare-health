```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# 1. Load the Data
# Ensure 'train.csv' is in the same directory
try:
    df = pd.read_csv('train.csv')
    print(f"Data loaded. Shape: {df.shape}")
except FileNotFoundError:
    print("Error: train.csv not found. Please ensure the file path is correct.")
    # For demonstration, assume 'df' is already loaded from your environment

# 2. Data Preprocessing
# 'datasetId' is likely an identifier and not a physiological feature, so we drop it.
X = df.drop(['condition', 'datasetId'], axis=1)
y = df['condition']

# Encode the target labels (e.g., 'no stress' -> 0, 'interruption' -> 1)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_names = le.classes_

# Split into Training (80%) and Testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Scale features to have a mean of 0 and variance of 1
# This helps models converge faster and perform better
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 3. Model Training
# We use 100 trees (n_estimators) and use all available CPU cores (n_jobs=-1)
print("Training Random Forest Model... (this may take a minute)")
model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_train_scaled, y_train)

# 4. Model Evaluation
y_pred = model.predict(X_test_scaled)

print(f"\nModel Accuracy: {accuracy_score(y_test, y_pred):.4%}")
print("\nDetailed Classification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))

# 5. Feature Importance Visualization
# This helps identify which HRV metrics are the best indicators of stress
importances = model.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': X.columns, 
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(15))
plt.title('Top 15 HRV Features for Stress Prediction')
plt.show()

# 6. Confusion Matrix Visualization
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Condition')
plt.ylabel('Actual Condition')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
import joblib  # Better than pickle for large numpy arrays
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder

# 1. Load Data
# Assuming your file is 'train.csv' based on previous snippet
df = pd.read_csv('train.csv')

# 2. Preprocessing
# We drop 'condition' (target) and 'datasetId' (non-physiological identifier)
X = df.drop(['condition', 'datasetId'], axis=1)
y = df['condition']

# Encode the target labels (e.g., 'no stress' -> 0)
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Train Model on the WHOLE dataset
print("Training the final model on all 369,289 rows...")
model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_scaled, y_encoded)

# 4. Bundle and Save to .pkl
# We save everything in a dictionary to keep them together
model_artifact = {
    'model': model,
    'scaler': scaler,
    'label_encoder': le,
    'feature_columns': list(X.columns)
}

joblib.dump(model_artifact, 'stress_prediction_model.pkl')
print("Successfully saved stress_prediction_model.pkl")
```
